```python
from pymongo import MongoClient
client = MongoClient()
client = MongoClient('mongodb://localhost:27017/')
db = client['application-task']
tweets = db.tweets

bagOfWords = ""
for tweet in tweets.find():
    for obj in tweet['hashtags']:
        bagOfWords += (obj['text']) + " "

bagOfWords[0:100]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

wordcloud = WordCloud(width=1200, height=1200, margin=0).generate(bagOfWords)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
